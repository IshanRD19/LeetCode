# DEEPEST LEAVES SUM

Given the root of a binary tree, return the sum of values of its deepest leaves.

### Example 1:

#### Input:
root = [1,2,3,4,5,null,6,7,null,null,null,null,8]
#### Output:
15
### Example 2:

#### Input:
root = [6,7,8,2,7,1,3,9,null,1,4,null,null,null,5]
#### Output:
19
 
### Constraints:
The number of nodes in the tree is in the range [1, 10^4].

1 <= Node.val <= 100

## Approach 1: Recursive DFS

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

# root: Optional[TreeNode]
# return -> int
class Solution:
    
    def dfs(self, root):
        
        # Check if root exists or not
        if not root:
            return 0
        
        # Count current node and move further down the tree
        return 1 + max(self.dfs(root.left), self.dfs(root.right))
    
    
    def deepest_sum(self, root, current_depth, max_depth):
        
        # Check if node exists
        if not root:
            return 0

        # Check if max depth is reached
        if current_depth == max_depth:
            return root.val
        
        return self.deepest_sum(root.left, current_depth + 1, max_depth) + self.deepest_sum(root.right, current_depth + 1, max_depth)
    
    
    def deepestLeavesSum(self, root):
        # Apply DFS to get maximum depth of tree
        max_depth = self.dfs(root) - 1
        
        return self.deepest_sum(root, 0, max_depth)
```

## Approach 2: BFS

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

# root: Optional[TreeNode]
# return -> int
class Solution:
    
    def deepestLeavesSum2(self, root):
        current_depth = [root]
        
        # Traverse until all nodes are explored
        while current_depth:
            current_sum = 0
            next_depth = []
            
            # For all nodes at current level
            for node in current_depth:
                
                # Update current depth sum
                current_sum += node.val
                
                # Gather nodes for next level
                if node.left:
                    next_depth.append(node.left)
                if node.right:
                    next_depth.append(node.right)
            
            # Move to next depth level
            current_depth = next_depth
            
        return current_sum
```
