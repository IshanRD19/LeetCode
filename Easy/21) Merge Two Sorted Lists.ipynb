# MERGE TWO SORTED LISTS

You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists in a one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

### Example 1:


#### Input:
list1 = [1,2,4], list2 = [1,3,4]
#### Output:
[1,1,2,3,4,4]
### Example 2:

#### Input:
list1 = [], list2 = []
#### Output:
[]
### Example 3:

#### Input:
list1 = [], list2 = [0]
#### Output:
[0]
 
### Constraints:
The number of nodes in both lists is in the range [0, 50].

-100 <= Node.val <= 100

Both list1 and list2 are sorted in non-decreasing order.

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# list1: Optional[ListNode]
# list2: Optional[ListNode]
# return -> Optional[ListNode]
class Solution:
    def mergeTwoLists(self, list1, list2):
        # Trivial checks
        if list1 and not list2:
            return list1
        elif list2 and not list1:
            return list2
        elif not list1 and not list2:
            return list1
        
        new_head, temp_node = None, None 
        
        # While both lists are not empty
        while list1 and list2:

            if not new_head:
            
                # Compare the elements for new head node
                if list1.val <= list2.val:
                    new_head = ListNode(list1.val)
                    list1 = list1.next
                    temp_node = new_head
                else:
                    new_head = ListNode(list2.val)
                    list2 = list2.next
                    temp_node = new_head

            # Compare the elements for following nodes
            else:
                if list1.val <= list2.val:
                    temp_node.next = ListNode(list1.val)
                    list1 = list1.next
                    temp_node = temp_node.next
                else:
                    temp_node.next = ListNode(list2.val)
                    list2 = list2.next
                    temp_node = temp_node.next
        
        if list1 and not list2:
            temp_node.next = list1

        elif list2 and not list1:
            temp_node.next = list2

        return new_head
```
